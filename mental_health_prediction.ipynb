```python
from sklearn.preprocessing import LabelEncoder

# To create an instance of LabelEncoder
label_encoder = LabelEncoder()

# Iterate over the columns of the DataFrame
for column in df.columns:
    # To check if the column data type is 'object' (categorical)
    if df[column].dtype == 'object':
        df[column] = label_encoder.fit_transform(df[column])
```

```python
df_shape = df.shape
print(df_shape)
```

```python
X = df.drop('mental_health_status', axis=1)
y = df['mental_health_status']
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.20, random_state=2)
```

```python
print("xtrain:", xtrain.shape)
print("xtest:", xtest.shape)
print("\nytrain:", ytrain.shape)
print("ytest:", ytest.shape)
```

```python
from sklearn.svm import SVC
model = SVC()

model.fit(xtrain, ytrain)

ytest_pred = model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(ytest, ytest_pred)
precision = precision_score(ytest, ytest_pred)
recall = recall_score(ytest, ytest_pred)
f1 = f1_score(ytest, ytest_pred)

print("Accuracy: {}".format(accuracy))
print("Precision: {}".format(precision))
print("Recall: {}".format(recall))
print("F1 Score: {}".format(f1))
```

```python
## Saving the trained SVM model
import joblib
joblib.dump(model, 'mental_health_recognition.pkl')
```

#### Calculating Feture ranges

```python
selected_features = [
    "anxiety_level",
    "depression",
    "sleep_quality",
    "stress_level",
    "social_support",
    "self_esteem",
    "future_career_concerns",
    "extracurricular_activities",
    "peer_pressure",
    "living_conditions"
]

# Getting min and max values for selected features
min_max_values = {feature: {"min": original_df[feature].min(), "max": original_df[feature].max()} for feature in selected_features}
```

```python
min_max_values
```

```python
import pandas as pd
feature_ranges = pd.DataFrame(columns=['Feature', 'Low', 'High'])

for feature, values in min_max_values.items():
    threshold = values['min'] + (values['max'] - values['min']) * 0.5

    low_category = f"({values['min']}-{int(threshold)})"
    high_category = f"({int(threshold)}-{values['max']}]"

    feature_ranges = pd.concat([feature_ranges, pd.DataFrame({
        'Feature': [feature.replace('_', ' ').title()],
        'Low': [low_category],
        'High': [high_category]
    })], ignore_index=True)

# Displaying the feature ranges DataFrame
feature_ranges.set_index('Feature', inplace=True)
print(feature_ranges)
```
